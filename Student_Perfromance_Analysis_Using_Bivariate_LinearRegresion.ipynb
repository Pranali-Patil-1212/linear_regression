```python
X = data.drop('Performance_Index', axis=1)
y = data['Performance_Index']
```

```python
X.head()
```

```python
y.head()
```

```python
X = pd.get_dummies(X, columns= ['Extracurricular_Activities'], drop_first= True)
X.head()
```

```python
X.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
print(X_train.shape)
print(X_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
from sklearn.metrics import r2_score
print("Train Accuracy:", r2_score(lr.predict(X_train), y_train))
print("Test Accuracy:", r2_score(lr.predict(X_test), y_test))
```

This model is good fit.
