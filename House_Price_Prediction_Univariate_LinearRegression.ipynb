# House Price Prediction
In this example, we are predicting the price of house based on the region. Here we are using univariate analysis. we are analyze the data using linear regression.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/All CSV_Files/house_prices.csv")
```

```python

plt.title('Area Box Plot')
sns.boxplot(x=data.area)
plt.show()
```

```python
plt.title("Price Box plot")
sns.boxplot(x=data.price)
plt.show()
```

```python
plt.title("Dist plot of price")
sns.distplot(x=data.price)
plt.show()
```

```python
plt.title("Scatter plot")
sns.scatterplot(x=data.area, y=data.price)
plt.show()
```

```python
X= data.drop("price", axis=1)
y=data["price"]
```

```python
X.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
print(X_train.shape, X_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
from sklearn.metrics import r2_score
print("Train Accuracy:", r2_score(lr.predict(X_train), y_train))
print("Test Accuracy:", r2_score(lr.predict(X_test), y_test))
```

This model is good fit because there is no difference between training and testing data and also both values are close to 1
